```python
# function to read file
def read_text_file(file_path):
    with open(file_path, 'r') as f:
        tex = list(f.read())
        return remove_values_from_list(tex,"\n"); # remove end line character

def remove_values_from_list(the_list, val):
   return [value for value in the_list if value != val]

def readFiles(path):
    # initialize lists
    data_j = [];
    data_s = [];
    data_e = [];
    # iterate through all file
    for file in os.listdir():
        # Check whether file is in text format or not
        if file.endswith(".txt"):
            file_path = f"{path}{file}"
            if file.startswith("j"):
                # call read text file function
                data_j.append(read_text_file(file_path))
            if file.startswith("e"):
                # call read text file function
                data_e.append(read_text_file(file_path))
            if file.startswith("s"):
                # call read text file function
                data_s.append(read_text_file(file_path))
    return data_j, data_s, data_e;

def countLetters(tex,alpha):
    cTypes = list("abcdefghijklmnopqrstuvwxyz ");
    c = [];
    for i in range(0,len(cTypes)):
        temp = alpha;
        for j in range(0,len(tex)):
            temp = temp+tex[j].count(cTypes[i]);
        c.append(temp);
    c = np.array(c);
    return c;

def computePx(count,Px):
    temp = 0;
    for i in range(0,len(count)):
        temp = temp+(math.log(Px[i])*count[i]);
    return temp;

def computePX_Y(theta,x):
    px_y = np.dot(np.log(theta),x);
    #px_y = math.exp(px_y);
    return px_y;

def computePy_x(count,Px,Px_y):
    Py_x = np.zeros([3,1]);
    for i in range(0,3):
        Py_x[i] = math.exp(math.log(1/3)+px_y[i]-pX10);
    return Py_x;

```
